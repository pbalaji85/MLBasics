# Home Prices Regression #
This is my attempt at predicting home sale prices using data from the [Home Prices: Advanced Regression Techniques](https://www.kaggle.com/c/house-prices-advanced-regression-techniques) kaggle contest. All of the data was downloaded from Kaggle. Let's get started!

```python
# initializing our libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
pd.options.display.max_columns = 100
```

```python
# Reading in the dataset
train = pd.read_csv('data/train.csv', index_col="Id")
print("The dimensions of the training dataset are: {}".format(train.shape))
test = pd.read_csv('data/test.csv', index_col="Id")
print("The dimensions of the test dataset are: {}".format(test.shape))
```

I am going to merge the training and test datasets together. This will help to understand the data better (since there will be more records in the combined dataset), and help in dealing with missing values, encoding categorical features and/or other data manipulation steps for both datasets at the same time.

```python
## Merge the training and test datasets together
mergedf = pd.concat([train.loc[:,'MSSubClass':'SaleCondition'],test.loc[:,'MSSubClass':'SaleCondition']])
print("The dimensions of the merged dataset are: {}".format(mergedf.shape))
mergedf.head()
```

This data has a large number of attributes, and several of them contain missing values. plotting can be a good way to visualize missing values

```python
def missing_values(df):
    ## Calculate percentage of missing value in each attribute
    na_prop = (df.isnull().sum()/len(df))*100
    ## Sort the series by value
    na_prop.sort_values(ascending=False, inplace=True)
    ## Convert the series to a pandas data frame
    na_prop = pd.DataFrame({'FeatureName':na_prop.index,'PercentMissValue':na_prop.values})
    return na_prop

miss_val = missing_values(mergedf)
```

```python
plt.figure(figsize=(15,7))
sns.set(style="ticks",context="talk")
sns.barplot(x = 'FeatureName',y = 'PercentMissValue',data = miss_val, palette = 'Paired')
plt.xticks(rotation=90)
plt.xlabel("Attributes")
plt.ylabel("% Missing Value")
plt.title("Percentage missing values by attribute (sorted)")
```

Let's go ahead and drop attributes that are largely composed of missing values. In this case they are the first 5 features in the above plot

```python
mergedf.drop(miss_val.FeatureName[0:5], axis = 1, inplace = True)
```

```python
print("The dimensions of the merged dataset with 5 features removed is: {}".format(mergedf.shape))
```

It will be interesting to see what features correlate most with home sale prices. To do this, I will have to go back to the training data set, which includes the home sale prices.

```python
cor = train.corr()['SalePrice'].copy()
cor.sort_values(ascending=False,inplace=True)
plt.figure(figsize=(10,12))
sns.set(style="ticks",context="talk")
sns.barplot(x=cor.values[1:len(cor.values)], y=cor.index[1:len(cor.index)], palette='Paired')
plt.xlabel("Features")
plt.ylabel("Correlation values")
plt.title("Pearson Correlation of Home Sale Prices with other Features")
```

A number of features appear to correlate positively with sale price. It will be interesting to plot a few of the highly correlated features against home prices. The most highly correlated feature is overall build and material quality.

```python
plt.figure(figsize=(12,6))
sns.set(style="ticks",context="talk")
sns.boxplot(x='OverallQual', y='SalePrice', data=train, palette='Paired')
plt.xlabel("Overall Build and Material Quality")
plt.ylabel("Home Sale Prices")
plt.title("Home Sale Prices by Overall Build and Material Quality")
```

As expected, the above plot shows us that home sale prices are strongly dictated by overall build and material quality. Higher scores are indicative of higher prices and vice versa.  
The second most correlated feature is 'GrLivArea', which is the above ground living area in square feet.

```python
plt.figure(figsize=(8,6))
sns.set(style="ticks",context="talk")
sns.regplot(y=train['SalePrice'], x=train['GrLivArea'], fit_reg=True, color='darkred', marker='.')
plt.xlabel("Above Ground Living Area (square feet)")
plt.ylabel("Home Sale Price")
plt.title("Correlation of Home Sale Prices with Above Ground Living Area")
```

Above ground living area shows a strong linear correlation with the home sale price.  
The third most correlated feature is 'GarageCars' - an indicator of the size of the garaze in number of cars and the fourth most correlated feature is 'GarageArea' - The area of the garage in square feet

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15,6), sharex=False,sharey=True)
ax1 = sns.boxplot(x='GarageCars', y='SalePrice', data=train, palette='Paired',ax=ax1)
ax1.set_xlabel('Garage Size (in Number of Cars)')
ax1.set_ylabel("Home Sale Prices")
ax2 = sns.regplot(x='GarageArea', y='SalePrice', data=train, fit_reg=True, marker='.', color='darkgreen')
ax2.set_xlabel("Garage Area (square feet)")
ax2.set_ylabel("")
plt.suptitle("Home Sale Price by Garage metrics")
```

The last feature that I'll plot before moving on is the fifth most correlated feature, i.e 'TotalBsmtSF' - Total square feet of basement area.

```python
plt.figure(figsize=(8,6))
sns.set(style="ticks",context="talk")
sns.regplot(y=train['SalePrice'], x=train['TotalBsmtSF'], fit_reg=True, color='darkblue', marker='.')
plt.xlabel("Total Basement Area (Square Feet)")
plt.ylabel("Home Sale Price")
plt.title("Correlation of Home Sale Prices with Total Basement Area (Square Feet)")
```

* The total basement area (TotalBsmtSF) is a sum of the type 1 finished area ('BsmtFinSF1'), type 2 finished area ('BsmntFinSF2'), and unfinished basement area ('BsmtUnfSF'). I am going to drop each of the individual columns while retaining the 'TotalBsmtSF' attribute.
* Similarly, the above ground living area ('GrLivArea') is a sum of the first floor ('1stFlrSF') and second floor ('2ndFlrSF') area. Again, all 3 features correate positively with home sale price, so I'm going to drop the individual areas while retaining the 'GrLivArea' attribute.

```python
mergedf.drop(['BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','1stFlrSF','2ndFlrSF'], axis=1, inplace=True)
```

Let's now go ahead and start dealing with missing values. To that end, let's identify and separate features that are numerical and those that are categorical.

```python
## numerical features
num = mergedf.select_dtypes(include='number').columns
## categorical features
cat = mergedf.select_dtypes(exclude='number').columns
```

Let's start the numerical values and fill in the missing values using the median of the column.

```python
## fill in missing values for the numerical columns
mergedf[num]=mergedf[num].fillna(value=mergedf[num].median(axis=0))
```

Before we finish with the numerical values, let's take a moment and check to see if any of the features are skewed (i.e they are asymmetrically distributed around a mean). Skewness of a feature adversely impacts building mathematical models. Machine learning models like data to be symetrically distributed around the mean.

```python
skewed = mergedf[num].skew(axis=0,skipna=True).sort_values(ascending=False)
skewed = pd.DataFrame({'FeatureName':skewed.index,'Skewness':skewed.values})
plt.figure(figsize=(15,10))
sns.set(style="ticks",context="talk")
sns.barplot(y=skewed['FeatureName'], x=skewed['Skewness'], palette='Paired')
```

From this plot, it is clear that a large number of features have a positive skew, which is easily fixed by log transformation. There are also some features with mild negative skew, which will be ignored for the purposes of this model. 

```python
skew_cols = skewed.FeatureName[(skewed['Skewness']>0.75)]
mergedf[skew_cols] = mergedf[skew_cols].apply(np.log1p, axis=0)
```

Now that we are done with the numerical features, let's encode all of the various categorical features in our dataset.

```python
mergedf = pd.get_dummies(mergedf, columns=cat)
```

Finally, before we get to the model building section, I am going to go ahead and split the 'mergedf' dataframe back into the respective training and test datasets

```python
final_train = mergedf.loc[1:len(train),]
final_test = mergedf.loc[(len(train)+1):len(mergedf),]
Y = train['SalePrice']
```

Now that the test and train data sets have been separated, it's time to move on generating regression models to predict home prices. Before we need to split the training dataset (to allow error estimation of the model)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_train, Y, test_size=0.33, random_state=42)
```

We're all set to build models. Let's start with ordinary squares linear regression. Given the rigidity of linear regression, it is expected that this model will yield the worst predictions. 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize=True)
lr.fit(X_train, y_train)
```

```python
## predict Y using linear regression model
lin_predict = lr.predict(X_test)
```

Now that we've predicted some home prices using the linear regression model, let's evaluate the accuracy of this model.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print('\nMetrics for the Linear Regression \n')
print("Mean Absolute Error (MAE):",mean_absolute_error(y_test, lin_predict))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, lin_predict)))
print("Median Home price:", Y.median())
```

The root mean squared error value for the linear regression is approximately $32,000, or an error margin of 20% of the median home value. If we predicted home values using this model, how would it fare on Kaggle?

```python
lin_results = pd.DataFrame(lr.predict(final_test),columns=['SalePrice'], index=test.index)
pd.DataFrame.to_csv(lin_results, path_or_buf='data/linear.csv')
```

This model doesn't fare particularly well on **Kaggle** with a score of **0.62106**.  
Perhaps regularized linear models will fare better than ordinary least squares linear regression. I am going to start with Lasso, which uses the l1-norm as a regularizer. I expect l1-norm to fare better (than the l2-norm) as the loss function minimizes the least absolute error.   
*Note:* I'm running Lasso using the default alpha value of 1.0. 

```python
from sklearn.linear_model import Lasso
lasso = Lasso(max_iter=10000, normalize=True, random_state=42)
lasso.fit(X_train, y_train)
```

```python
las_predict = lasso.predict(X_test)
```

Let's evaluate this model to see if the Lasso produces better results than ordinary least squares regression.

```python
print('\nMetrics for the Lasso Regression \n')
print("Mean Absolute Error (MAE):",mean_absolute_error(y_test, las_predict))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, las_predict)))
```

The RMSE values for Lasso are certainly marginally better than ordinary least squares regression. Let's see how the results from Lasso fare on **Kaggle**.

```python
las_results = pd.DataFrame(lasso.predict(final_test),columns=['SalePrice'], index=test.index)
pd.DataFrame.to_csv(las_results, path_or_buf='data/lasso.csv')
```

As one would expect, based on the RMSE values for Lasso, it fared slightly better than linear regression with a **Kaggle** score of **0.61707**.  
Can Ridge regression with its l2-norm regularization perform better than the Lasso? 

```python
from sklearn.linear_model import Ridge
rid = Ridge(max_iter=10000, random_state=42, normalize=True)
rid.fit(X_train, y_train)
```

```python
rid_predict = rid.predict(X_test)
```

Let's see how the l2-norm regularization of Ridge performs in predicting home prices, as opposed to Lasso

```python
print('\nMetrics for the Ridge Regression \n')
print("Mean Absolute Error (MAE):",mean_absolute_error(y_test, rid_predict))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, rid_predict)))
```

Let's see how the ridge regression scores on home sale price prediction on **Kaggle**

```python
rid_results = pd.DataFrame(rid.predict(final_test),columns=['SalePrice'], index=test.index)
pd.DataFrame.to_csv(rid_results, path_or_buf='data/ridge.csv')
```

Despite the higher RMSE values, the Ridge actually scores better with a score of **0.15319** on **Kaggle**.  

Let's see how cross-validated models performs with this dataset. The advantage of cross-validation is that it cycles through a large number of values (default is 100), to find the best alpha for the dataset. Cross-validation works better when you give the model the entire training dataset (instead of hiding away a fraction). The risk with cross-validation is that it may over-fit to the training dataset. Let's start with cross-validated Lasso.

```python
from sklearn.linear_model import LassoCV
lascv = LassoCV(n_alphas=1000, max_iter=10000, n_jobs=8, normalize=True)
lascv.fit(final_train, Y)
```

```python
lascv_predict = lascv.predict(X_test)
```

```python
print('\nMetrics for the LassoCV \n')
print("Mean Absolute Error (MAE):",mean_absolute_error(y_test, lascv_predict))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, lascv_predict)))

## Predict home sale prices using the LassoCV model.
lassocv_results = pd.DataFrame(lascv.predict(final_test),columns=['SalePrice'], index=test.index)
pd.DataFrame.to_csv(lassocv_results, path_or_buf='data/lassocv.csv')
```

That is the lowest RMSE value thus far. **On Kaggle**, the LassoCV model scores **0.42549**.  
How would cross-validated Ridge perform? Let's submit predictions to Kaggle and find out.

```python
from sklearn.linear_model import RidgeCV
rid_cv = RidgeCV(alphas=(0.1,0.5,1,5,10,50,100,500), cv=1000, normalize=True)
rid_cv.fit(final_train, Y)
```

```python
ridcv_predict = rid_cv.predict(X_test)
```

```python
print('\nMetrics for the RidgeCV \n')
print("Mean Absolute Error (MAE):",mean_absolute_error(y_test, ridcv_predict))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, ridcv_predict)))

## Predict home sale prices for the test dataset
ridgecv_results = pd.DataFrame(rid_cv.predict(final_test),columns=['SalePrice'], index=test.index)
pd.DataFrame.to_csv(ridgecv_results, path_or_buf='data/ridgecv.csv')
```

Cross-validated Ridge Regression scores remarkably well on **Kaggle** with a score of **0.14814**.

Let's see how ensemble models (Random Forest) perform compared to cross-validated regularized models. I am going to build the model, predict home sale prices for the test dataset and score the predictions on Kaggle. 

```python
from sklearn.ensemble import RandomForestRegressor
random = RandomForestRegressor(random_state=42)
random.fit(X_train, y_train)
```

```python
rand_predict = random.predict(X_test)
```

```python
print('\nMetrics for the Random Forest Regression \n')
print("Mean Absolute Error (MAE):",mean_absolute_error(y_test, rand_predict))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, rand_predict)))

## Make predictions for the test dataset
random_results = pd.DataFrame(random.predict(final_test),columns=['SalePrice'], index=test.index)
pd.DataFrame.to_csv(random_results, path_or_buf='data/random.csv')
```

The Random Forest actually performs very well with a **Kaggle** score of **0.16162**. The model with the best score thus far has been the cross-validated Ridge.  
Will the Random Forest predictions get better if we perforned hyperparameter tuning using Gridsearch?

```python
from sklearn.model_selection import GridSearchCV

## specify the parameter grid for hyperparameter tuning
param_grid = [
    {'n_estimators': [3, 10, 30,40,50,60,70,80,100], 'max_features': [2, 4, 6, 8,10,30,40,50,60]},
    {'bootstrap': [False], 'n_estimators': [3, 10,30,40,50,80,100], 'max_features':[2, 3, 4,10,30,40,50,60]},
]

gridsearch = GridSearchCV(random, param_grid, cv=100, scoring='neg_mean_squared_error', n_jobs=16)
gridsearch.fit(final_train, Y)
```

```python
gridsearch.best_params_
```

```python
## Predict the test dataset and evaluate results on Kaggle
gridsearch_results = pd.DataFrame(gridsearch.predict(final_test),columns=['SalePrice'], index=test.index)
pd.DataFrame.to_csv(gridsearch_results, path_or_buf='data/gridsearch.csv')
```

Hyperparameter tuning gives the best predictions with a **Kaggle** score of **0.14738**.
